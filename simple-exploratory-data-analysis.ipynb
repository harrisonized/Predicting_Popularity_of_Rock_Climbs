```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
import math
import warnings # Turn off warnings
warnings.filterwarnings('ignore')
```

```python
def plot_heat_map(df):
    global ax
    plt.figure(figsize=(8, 5))
    sns.heatmap(df.corr(), cmap= 'coolwarm', annot=True)
    ax = plt.axes()

def plot_scatter(x, y, color):
    global ax
    plt.figure(figsize=(8, 5))
    plt.scatter(x, y, color=color)
    ax = plt.axes()
```

```python
"""
Note: Exports are hashed out.
"""

bishop_proc_df = pd.read_csv(r"Data/Processed CSV/bishop_proc_df.csv")
bishop_proc_df = bishop_proc_df[["Avg Stars", "Grade", "Length", "StarRatings", "OnToDoLists", "Ticks"]].dropna()
```

```python
# Pairplot
features_target_list = ['StarRatings', 'Ticks', 'Avg Stars', 'Length', 'Grade', 'OnToDoLists']
sns.pairplot(bishop_proc_df[features_target_list])
#plt.savefig("Figures/Bishop/EDA/Baseline Pairplot.png", transparent=True, bbox_inches="tight")
```

```python
# Plot heatmap for baseline case.
plot_heat_map(bishop_proc_df.loc[:,['StarRatings', 'Ticks', 'Avg Stars', 'Length', 'Grade', 'OnToDoLists']])
ax.set_title('Correlation Heatmap for True Baseline Case', fontsize = 18)
#plt.savefig("Figures/Bishop/EDA/Correlation Heatmap.png", transparent=True, bbox_inches="tight")
```

```python
# Scatterplots that are the last row of the pairplot

plot_scatter(bishop_proc_df['StarRatings'], bishop_proc_df['OnToDoLists'], '#4568c2')
ax.set_title('OnToDoLists vs StarRatings', fontsize = 18)
ax.set_xlabel('StarRatings', fontsize=16)
ax.set_xlim(left = -45, right = 845)
ax.set_ylabel('OnToDoLists', fontsize=16)
ax.set_ylim(bottom = -45, top = 645)
plt.savefig("Figures/Bishop/EDA/OnToDoLists vs StarRatings.png", transparent=True, bbox_inches="tight")

plot_scatter(bishop_proc_df['Ticks'], bishop_proc_df['OnToDoLists'], '#8dde19')
ax.set_title('OnToDoLists vs Ticks', fontsize = 18)
ax.set_xlabel('Ticks', fontsize=16)
ax.set_xlim(left = -45, right = 845)
ax.set_ylabel('OnToDoLists', fontsize=16)
ax.set_ylim(bottom = -45, top = 645)
plt.savefig("Figures/Bishop/EDA/OnToDoLists vs Ticks.png", transparent=True, bbox_inches="tight")

plot_scatter(bishop_proc_df['Avg Stars'], bishop_proc_df['OnToDoLists'], "#f0cb2f")
ax.set_title('OnToDoLists vs Avg Stars', fontsize = 18)
ax.set_xlabel('Avg Stars', fontsize=16)
ax.set_xlim(left = -0.25, right = 4.5)
ax.set_ylabel('OnToDoLists', fontsize=16)
ax.set_ylim(bottom = -45, top = 645)
plt.savefig("Figures/Bishop/EDA/OnToDoLists vs Avg Stars.png", transparent=True, bbox_inches="tight")

plot_scatter(bishop_proc_df['Length'], bishop_proc_df['OnToDoLists'], "#ffa500")
ax.set_title('OnToDoLists vs Length', fontsize = 18)
ax.set_xlabel('Length', fontsize=16)
ax.set_xlim(left = -4.5, right = 64.5)
ax.set_xticks(np.arange(0, 64.5, 5))
ax.set_ylabel('OnToDoLists', fontsize=16)
ax.set_ylim(bottom = -45, top = 645)
plt.savefig("Figures/Bishop/EDA/OnToDoLists vs Length.png", transparent=True, bbox_inches="tight")

plot_scatter(bishop_proc_df['Grade'], bishop_proc_df['OnToDoLists'], color='#666666')
ax.set_title('OnToDoLists vs Grade', fontsize = 18)
ax.set_xlabel('Grade', fontsize=16)
ax.set_xticks(np.arange(-1, 15, 1))
ax.set_xlim(left = -2, right = 15)
ax.set_ylabel('OnToDoLists', fontsize=16)
plt.savefig("Figures/Bishop/EDA/OnToDoLists vs Grade.png", transparent=True, bbox_inches="tight")

# Histogram on 'OnToDoLists'
plt.figure(figsize=(8, 5))
ax = plt.axes()
ax.hist(bishop_proc_df['OnToDoLists'], bins=np.arange(0, 201, 10), color='#8d1a93')
ax.set_title('OnToDoLists Histogram', fontsize = 18)
ax.set_xlabel('OnToDoLists', fontsize=16)
ax.set_ylabel('Count', fontsize=16)
ax.set_xlim(left = -1, right=150)
ax.set_xticks(np.arange(0, 201, 50))
plt.savefig("Figures/Bishop/EDA/OnToDoLists Histogram.png", transparent=True, bbox_inches="tight")
```
