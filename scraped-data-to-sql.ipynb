```python
import pandas as pd
import math

import pandas.io.sql as pd_sql
import psycopg2 as pg
from sqlalchemy import create_engine

# psychopg connection parameters
params = {'user': 'postgres', 'host': 'localhost', 'port': 5432, 'password': None}
connection = pg.connect(**params, dbname='mountain_project') # Connect
cursor = connection.cursor()

# sqlalchemy connection arguments
connection_str = f'postgres://postgres:{None}@localhost:5432/mountain_project'
engine = create_engine(connection_str)
```

```python
buttermilks_scrape_df = pd.read_csv(r"data/scraped-data/buttermilks-scrape.csv")
buttermilks_scrape_df[['url', 'star_ratings', 'suggested_ratings', 'on_to_do_lists', 'ticks']].to_sql('buttermilks_scrape', engine, index=True) # Create the database

druid_stones_scrape_df = pd.read_csv(r"data/scraped-data/druid_stones-scrape.csv")
druid_stones_scrape_df[['url', 'star_ratings', 'suggested_ratings', 'on_to_do_lists', 'ticks']].to_sql('druid_stones_scrape', engine, index=True) # Create the database

happy_boulders_scrape_df = pd.read_csv(r"data/scraped-data/happy_boulders-scrape.csv")
happy_boulders_scrape_df[['url', 'star_ratings', 'suggested_ratings', 'on_to_do_lists', 'ticks']].to_sql('happy_boulders_scrape', engine, index=True) # Create the database

sad_boulders_scrape_df = pd.read_csv(r"data/scraped-data/sad_boulders-scrape.csv")
sad_boulders_scrape_df[['url', 'star_ratings', 'suggested_ratings', 'on_to_do_lists', 'ticks']].to_sql('sad_boulders_scrape', engine, index=True) # Create the database

joshua_tree_scrape_df = pd.read_csv(r"data/scraped-data/joshua_tree-scrape.csv")
joshua_tree_scrape_df[['url', 'star_ratings', 'suggested_ratings', 'on_to_do_lists', 'ticks']].to_sql('joshua_tree_scrape', engine, index=True) # Create the database
```
